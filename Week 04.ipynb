## While Loop

```python
i=0
while i<5:
    print(i)
    i+=1
```

```python
password = ""
while password != "python123":
    password = input("Enter the Password: ")

print("Access Granted")
```

```python
import time
battery=10
while battery>0:
    print("Battery Level:",battery)
    battery-=1
    time.sleep(1.5)
print("Battery Empty")
```

```python
age=-1
while age<0:
    age-int(input("Enter a valid age : "))
    print ("Thank You")
    break
```

```python
choice=""
while choice!="3":
    print("1.Add\n2.View\n3.exit")
    choice=input("Choose an option :")
    print ("GoodBye!")
    break
```

### (\n) - Endline Characters

```python
for i in range(2,5):
    for j in range(1,i):
        print(j,end="")
    print()
```

```python
count=5
while count>0:
    print(count,end="")
    count-=2
```

```python
first_name = input("Enter first name: ")
second_name = input("Enter second name: ")
age = int(input("Enter the age: "))

print(f"My First Name is: {first_name}, Second Name: {second_name}, Age: {age}")
```
